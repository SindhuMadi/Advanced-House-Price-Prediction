```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import Lasso, Ridge
from sklearn.feature_selection import SelectFromModel

%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('X_train.csv')
```

```python
data.head()
```

```python
# Dependent feature
y_train = data['SalePrice']
```

```python
X_train = data.drop(['Id','SalePrice'], axis=1)
```

```python
X_train.head()
```

```python
## Feature Selection
# Lasso regression model by selecting a suitable alpha. The bigger the alpha, lesser features will be selected

#We use SelectFromModel object from sklearn, which automatically selects features with non-zero coefficients

feat_sel_model = SelectFromModel(Ridge(alpha = 0.001, random_state=0))
feat_sel_model.fit(X_train, y_train)
```

```python
feat_sel_model.get_support()  #True - important feature, False - not important feature ans can be skipped
```

```python
#Printing the total and selected features
selected_feat = X_train.columns[feat_sel_model.get_support()]

print("Total Features: {}".format((X_train.shape[-1])))
print("Selected Features: {}".format(len(selected_feat)))
print("Features with coefficients shrank to zero: {}".format(np.sum(feat_sel_model.estimator_.coef_==0)))
```

```python
selected_feat
```

```python
X_train[selected_feat].head()
```

```python
X_train = X_train[selected_feat]
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor(n_estimators = 250, random_state=0)
model2.fit(X_train, y_train)
```

```python
model2.score(X_train, y_train)
```

### XGBOOST

```python
import xgboost
regressor=xgboost.XGBRegressor()
```

```python
booster=['gbtree','gblinear']
base_score=[0.25,0.5,0.75,1]
```

```python
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=2)
```

```python
random_cv.fit(X_train,y_train)
```
